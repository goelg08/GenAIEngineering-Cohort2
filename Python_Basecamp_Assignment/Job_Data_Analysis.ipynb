```python
import pandas as pd
```

```python
class Job_Data_Analysis:
    
    def __init__(self, file_name):
        self.file_name = file_name
        
    def get_company_locations(self):
        job_dataset = pd.read_csv(self.file_name)
        return job_dataset['company_location']
    
    def get_salary_range_per_emp_type(self, country):
        job_dataset = pd.read_csv(self.file_name)
        filtered_job_dataset = job_dataset[job_dataset['company_location'] == 'China']
        group_by_dataset = filtered_job_dataset.groupby ('employment_type') ['salary_usd'].agg(['min', 'max', 'mean'])
        return group_by_dataset
    
    def get_avg_exp_per_level(self, country):
        job_dataset = pd.read_csv(self.file_name)
        filtered_job_dataset = job_dataset[job_dataset['company_location'] == 'China']
        group_by_dataset = filtered_job_dataset.groupby ('experience_level') ['years_experience'].agg(['mean'])
        return group_by_dataset
    
    def get_num_industry(self):
        job_dataset = pd.read_csv(self.file_name)
        group_by_dataset = job_dataset.groupby ('company_location') ['industry'].agg(['count'])
        return group_by_dataset
    
    def get_benefit_score_range(self):
        job_dataset = pd.read_csv(self.file_name)
        group_by_dataset = job_dataset.groupby ('company_location') ['benefits_score'].agg(['min', 'max', 'mean'])
        return group_by_dataset
    
```

```python
Data_Analyst_1 = Job_Data_Analysis(file_name="ai_job_dataset.csv")
comp_loc = Data_Analyst_1.get_company_locations().tolist()
print("List of Company Location : ", comp_loc)

sal_range_per_emp_dataset = Data_Analyst_1.get_salary_range_per_emp_type(country="China")
sal_range_per_emp_dataset

avg_exp_per_level_dataset = Data_Analyst_1.get_avg_exp_per_level(country="China")
dict_avg_per_level = avg_exp_per_level_dataset.to_dict()
print(dict_avg_per_level)


num_industries_per_country = Data_Analyst_1.get_num_industry()
dict_num_per_country = num_industries_per_country.to_dict()
print(dict_num_per_country)

ben_score_per_country_dataset = Data_Analyst_1.get_benefit_score_range()
ben_score_per_country_dataset
```
